# MSA-NARX Quantile Pipeline — Calibrated Uncertainty (version_4_quantile)

Adds a **monotonic 3-quantile head** (τ=[0.05,0.50,0.95]) to the MSA-NARX point model (ver_3).
Output **(B, M=15, 26, 3)** = lower / median / upper per output per horizon — drop-in for the
common CQR pipeline (`calibrate_cqr` / `calibrate_cqr_ar`).

- Warm-started from `Models_MSA/version_3` (median == ver_3 at init); bands learn via PinballLoss.
- Same recipe as ver_3 (lr=1e-4, grad-clip 0.5, noise σ=0.05) — avoided gradient explosion.
- ver_3 is **not** modified. Saved to `Models_MSA/version_4_quantile/`.
